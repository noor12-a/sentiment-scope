```python
def to_lower(text: str) -> str:
    """Convert text to lower‑case.

    Args:
        text: Input string.
    Returns:
        Lower‑cased string.
    """
    return text.lower()

def remove_urls(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)

def remove_html_tags(text):
    return re.sub(r'<.*?>', '', text)
result = remove_html_tags("<p>This product is <b>amazing</b> and worth it.</p>")   
print("3.Remove HTML Tags:")
print (result)

def remove_punctuation(text: str) -> str:
    """Remove punctuation characters.

    Uses ``string.punctuation`` which covers the most common ASCII symbols.
    """
    return text.translate(str.maketrans("", "", string.punctuation))


def remove_numbers(text: str) -> str:
    """Strip all numeric characters from the string.
    """
    return re.sub(r"\d+", "", text)


def strip_whitespace(text: str) -> str:
    """Collapse multiple whitespace characters into a single space and trim.
    """
    return " ".join(text.split())


def tokenize(text: str) -> List[str]:
    """Very simple whitespace tokenizer.
    """
    return text.split()


```

```python
stemmer = PorterStemmer()
def apply_stemming(text):
    tokens = text.split()
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    return " ".join(stemmed_tokens)
result = apply_stemming("running runner runs")    
print("8.Apply Stemming:")
print(result)
```

```python
def preprocess(text: str) -> str:
    """Apply the full preprocessing pipeline.

    The order is:
    1. lower‑case
    2. remove punctuation
    3. remove numbers
    4. strip extra whitespace
    """
    text = to_lower(text)
    text = remove_punctuation(text)
    text = remove_numbers(text)
    text = strip_whitespace(text)
    text = strip_whitespace(text)
    tokens = tokenize(text)
    text = " ".join(tokens)
    return text
```
